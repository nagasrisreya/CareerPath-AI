```python
def preprocess_image(image, target_size=(256, 256), apply_filter=True):
    """
    Resizes, normalizes, and optionally filters an image.

    Args:
        image (np.array): The input image (BGR format).
        target_size (tuple): Desired (width, height) for the resized image.
        apply_filter (bool): Whether to apply a Gaussian filter.

    Returns:
        np.array: The preprocessed image.
    """
    # Resize image
    resized_image = cv2.resize(image, target_size, interpolation=cv2.INTER_AREA)

    # Normalize pixel values to [0, 1]
    normalized_image = resized_image.astype('float32') / 255.0

    # Apply Gaussian filter for noise reduction (optional)
    if apply_filter:
        filtered_image = cv2.GaussianBlur(normalized_image, (5, 5), 0)
    else:
        filtered_image = normalized_image

    return filtered_image

# Apply preprocessing to our images
if 'img1' in locals() and img1 is not None and 'img2' in locals() and img2 is not None:
    img1_preprocessed = preprocess_image(img1)
    img2_preprocessed = preprocess_image(img2)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor((img1_preprocessed * 255).astype(np.uint8), cv2.COLOR_BGR2RGB))
    plt.title('Preprocessed Image 1')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor((img2_preprocessed * 255).astype(np.uint8), cv2.COLOR_BGR2RGB))
    plt.title('Preprocessed Image 2')
    plt.axis('off')
    plt.show()
    print("Images preprocessed successfully.")
else:
    print("Skipping preprocessing: Images were not loaded successfully.")
```

```python
if 'img1_preprocessed' in locals() and img1_preprocessed is not None:
    # Convert preprocessed images to grayscale for HOG
    img1_gray = cv2.cvtColor((img1_preprocessed * 255).astype(np.uint8), cv2.COLOR_BGR2GRAY)
    img2_gray = cv2.cvtColor((img2_preprocessed * 255).astype(np.uint8), cv2.COLOR_BGR2GRAY)

    # Extract HOG features and visualization image
    fd1, hog_image1 = hog(img1_gray, orientations=9, pixels_per_cell=(8, 8),
                        cells_per_block=(2, 2), visualize=True, channel_axis=None)
    fd2, hog_image2 = hog(img2_gray, orientations=9, pixels_per_cell=(8, 8),
                        cells_per_block=(2, 2), visualize=True, channel_axis=None)

    # Rescale histogram for better display
    hog_image_rescaled1 = exposure.rescale_intensity(hog_image1, in_range=(0, 10))
    hog_image_rescaled2 = exposure.rescale_intensity(hog_image2, in_range=(0, 10))

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(hog_image_rescaled1, cmap='gray')
    plt.title('HOG Features Image 1')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(hog_image_rescaled2, cmap='gray')
    plt.title('HOG Features Image 2')
    plt.axis('off')
    plt.show()

    print(f"HOG features extracted. Feature descriptor size for image 1: {fd1.shape}")
    print(f"HOG features extracted. Feature descriptor size for image 2: {fd2.shape}")
else:
    print("Skipping HOG feature extraction: Preprocessed images not available.")
```

```python
if 'img1' in locals() and img1 is not None and 'img2' in locals() and img2 is not None:
    # Initialize ORB detector
    orb = cv2.ORB_create(nfeatures=1000) # Increased nfeatures for potentially more keypoints

    # Find the keypoints and descriptors with ORB
    kp1, des1 = orb.detectAndCompute(cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY), None)
    kp2, des2 = orb.detectAndCompute(cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY), None)

    # Draw keypoints on the images
    img_kp1 = cv2.drawKeypoints(img1, kp1, None, color=(0, 255, 0), flags=0)
    img_kp2 = cv2.drawKeypoints(img2, kp2, None, color=(0, 255, 0), flags=0)

    plt.figure(figsize=(14, 7))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_kp1, cv2.COLOR_BGR2RGB))
    plt.title('ORB Keypoints Image 1')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(img_kp2, cv2.COLOR_BGR2RGB))
    plt.title('ORB Keypoints Image 2')
    plt.axis('off')
    plt.show()

    print(f"ORB keypoints found for image 1: {len(kp1)}")
    print(f"ORB descriptors shape for image 1: {des1.shape if des1 is not None else 'None'}")
    print(f"ORB keypoints found for image 2: {len(kp2)}")
    print(f"ORB descriptors shape for image 2: {des2.shape if des2 is not None else 'None'}")

    if des1 is None or des2 is None:
        print("Warning: Descriptors could not be computed for one or both images. This might happen if no keypoints were detected or due to image content.")
else:
    print("Skipping SIFT/ORB feature extraction: Images not loaded successfully.")
```

```python
if 'img1' in locals() and img1 is not None and 'img2' in locals() and img2 is not None and \
   'kp1' in locals() and kp1 is not None and 'des1' in locals() and des1 is not None and \
   'kp2' in locals() and kp2 is not None and 'des2' in locals() and des2 is not None:

    if des1 is not None and des2 is not None and len(kp1) > 0 and len(kp2) > 0:
        # Create BFMatcher object
        bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

        # Match descriptors
        matches = bf.match(des1, des2)

        # Sort them in the order of their distance (best matches first)
        matches = sorted(matches, key=lambda x: x.distance)

        # Draw top N matches
        num_matches_to_draw = min(50, len(matches)) # Draw at most 50 matches
        img_matches = cv2.drawMatches(img1, kp1, img2, kp2, matches[:num_matches_to_draw], None,
                                      flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

        plt.figure(figsize=(15, 10))
        plt.imshow(cv2.cvtColor(img_matches, cv2.COLOR_BGR2RGB))
        plt.title(f'Top {num_matches_to_draw} Feature Matches (ORB)')
        plt.axis('off')
        plt.show()

        print(f"Total matches found: {len(matches)}")
        print(f"Displayed top {num_matches_to_draw} matches.")
    else:
        print("Skipping feature matching: Descriptors or keypoints not available for matching. Ensure images were loaded and ORB features extracted correctly.")
else:
    print("Skipping feature matching: Images or ORB results not available. Please run previous cells.")
```

```python
def calculate_iou(boxA, boxB):
    # boxA and boxB are arrays like [x1, y1, x2, y2] (top-left, bottom-right corners)

    # Determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    # Compute the area of intersection rectangle
    inter_area = max(0, xB - xA) * max(0, yB - yA)

    # Compute the area of both the prediction and ground-truth rectangles
    boxA_area = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxB_area = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])

    # Compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the intersection area
    iou = inter_area / float(boxA_area + boxB_area - inter_area)

    return iou

print("The `calculate_iou` function has been defined.")
print("To evaluate performance, you would need ground truth bounding boxes and class labels for comparison.")
print("A conceptual `evaluate_model_performance` function would iterate through predictions and ground truths, calculate IoU, and then determine metrics like True Positives, False Positives, False Negatives, Precision, Recall, and Classification Accuracy.")
```

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model
# Using 'yolov8n.pt' (nano version) for faster inference
model = YOLO('yolov8n.pt')

# Ensure 'yolo input.jpg' is loaded as yolo_input_img
# If it was previously loaded, it should be in the kernel state.
# If not, you might need to re-run the cell that loads `yolo_input_img`.
image_path_for_yolo = 'yolo input.jpg' # Corrected path to match the uploaded file
yolo_input_img = cv2.imread(image_path_for_yolo)

if yolo_input_img is None:
    print(f"Error: Image not found at {image_path_for_yolo}. Please upload it or check the path.")
else:
    print(f"Performing object detection on YOLO input image: {image_path_for_yolo}...")
    # Perform inference
    # The 'conf' argument sets the object confidence threshold
    results = model(yolo_input_img, conf=0.25) # Lowered confidence threshold

    # Display results
    for r in results:
        im_array = r.plot()  # plot a BGR numpy array of predictions
        im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB) # Convert to RGB for matplotlib

        plt.figure(figsize=(10, 8))
        plt.imshow(im_rgb)
        plt.title(f'Object Detection with YOLOv8 on {image_path_for_yolo}')
        plt.axis('off')
        plt.show()

        print(f"Detected {len(r.boxes)} objects.")
        if len(r.boxes) > 0:
            for i, box in enumerate(r.boxes):
                cls = int(box.cls[0])
                conf = float(box.conf[0])
                # Get class name from model's class names mapping
                class_name = model.names[cls]
                print(f"Object {i+1}: Class='{class_name}', Confidence={conf:.2f}")
        else:
            print("No objects detected with the given confidence threshold.")
```
